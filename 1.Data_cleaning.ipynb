```python
import pandas as pd
```

```python

df = pd.read_excel(r"D:\Guvi\Project 1\traffic_stops.xlsx")
df.head()
```

```python
print(df.shape)      
print(df.columns)    
print(df.info())     
```

```python
#drop null columns
df.dropna(axis=1, how='all', inplace=True)
```

```python
df['driver_age'].fillna(df['driver_age'].median(), inplace=True)
df['violation'].fillna('Unknown', inplace=True)
df['search_type'].fillna('Unknown', inplace=True)
df['search_conducted'].fillna(False, inplace=True)
df['is_arrested'].fillna(False, inplace=True)
df['drugs_related_stop'].fillna(False, inplace=True)
```

```python
#drop duplicates
df.drop_duplicates(inplace=True)
```

```python
# convert stop_time strings to time
df['stop_time'] = pd.to_datetime(df['stop_time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
df['stop_date'] = pd.to_datetime(df['stop_date'], errors='coerce')
```

```python
df.drop(columns=['driver_age_raw'], inplace=True)
df.drop(columns=['violation_raw'], inplace=True)
df.drop(columns=['is_arrested'], inplace=True)
df.drop(columns=['drugs_related_stop'], inplace=True)
```

```python
print(df.columns)
```

```python
#export
df.to_csv("cleaned_traffic_stops.csv", index=False)
print("Saved cleaned_traffic_stops.csv")
```
