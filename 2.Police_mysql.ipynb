Creating Table

```python
# Create new database
mycursor.execute("CREATE DATABASE IF NOT EXISTS police_checkpost")
print("Database created or already exists!")

mycursor.execute("USE police_checkpost")
mycursor.execute("""
CREATE TABLE IF NOT EXISTS traffic_stops (
    id INT AUTO_INCREMENT PRIMARY KEY,
    stop_date DATE,
    stop_time TIME,
    country_name VARCHAR(100),
    driver_gender VARCHAR(10),
    driver_age INT,
    driver_race VARCHAR(50),
    violation VARCHAR(100),
    search_conducted BOOLEAN,
    search_type VARCHAR(100),
    stop_outcome VARCHAR(50),
    stop_duration VARCHAR(20),
    vehicle_number VARCHAR(20)
)
""")

print("Table 'traffic_stops' created successfully!")
```

Inserting Data

```python
import mysql.connector
import pandas as pd

df = pd.read_csv("cleaned_traffic_stops.csv")

myconnection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="mysql@007",
    database="police_checkpost"
)
cursor = myconnection.cursor()

insert_query = """
INSERT INTO traffic_stops (
    stop_date, stop_time, country_name, driver_gender, driver_age,
    driver_race, violation, search_conducted, search_type,
    stop_outcome, stop_duration, vehicle_number
)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""


for i in range(len(df)):

    values = tuple(x.item() if hasattr(x, "item") else x for x in df.iloc[i])
    cursor.execute(insert_query, values)

myconnection.commit()

cursor.close()
myconnection.close()
```
